```python
from tensorflow.keras.utils import load_img, img_to_array
```

```python
img_path = 'NonDendrictic.jpg'

img = load_img(img_path, target_size=(256, 256))
imshow(img)

x = img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
p = model.predict(x)
print(p)
if (p == 1):
    print("Its dendritic")
else:
    print("Its non dendritic")
```
